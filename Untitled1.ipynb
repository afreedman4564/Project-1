```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
spoon_pd = pd.read_csv("recipe_stats.csv")
spoon_pd.head()
```

```python
spoon_pd["name lower"] = spoon_pd["Name"].str.lower()
spoon_pd.head()
```

```python
#lists = ["beef_name_list", "fish_name_list", "poultry_name_list"]
beef_name_list = ["beef", "steak", "hamburger", "chuck"]
fish_name_list = ["fish", "salmon", "tuna", "tilapia", "cod"]
poultry_name_list = ["turkey", "chicken", "duck"]
spoon_pd["meat type"] = ""

def meat_type(meat_name_list, value):
    for word in meat_name_list:
        spoon_pd["meat type"] = np.where(spoon_pd["name lower"].str.contains(word), value, spoon_pd["meat type"])
        return spoon_pd["meat type"]

# for lst in lists:
#     for word in lst:
#         spoon_pd[value] = np.where(spoon_pd["name lower"].str.contains(word), value, "other")

meat_type("beef_name_list", "beef")
meat_type("fish_name_list", "fish")
meat_type("poultry_name_list", "poultry")
spoon_pd["meat type"].value_counts()
                            
#df['New Group'] = np.where(df['Old Group'].str.contains("Stark"), 'Stark Family', 'Other')
```

```python
y_axis_list = ["Cost", "Protein (g)", "Calories", "Fat (g)", "Servings", "Preparation Time (m)", "Health Score"]
x_axis_list = ["Cost", "Protein (g)", "Calories", "Fat (g)", "Servings", "Preparation Time (m)", "Health Score"]

for y in y_axis_list:
    for x in x_axis_list:
        if x != y:
            plt.scatter(spoon_pd[x], spoon_pd[y], edgecolor = "black")
            (slope, intercept, rvalue, pvalue, stderr) = st.linregress(spoon_pd[x], spoon_pd[y])
            regress_values = intercept + slope*spoon_pd[x]
            r_squared_value = f"r-squared: {round(rvalue**2, 2)}"
            plt.ylabel(f"{y}")
            plt.xlabel(f"{x}")
            plt.title(f"{x} Vs. {y}")
            plt.figtext(1.05, .42, r_squared_value, color = "r")
            plt.subplots_adjust(bottom = 0.35, right = 1)
            plt.show()
```
